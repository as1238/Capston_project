```python
DPI_prices = pd.read_csv(path_to_module+ '/Historical_Token_Prices.csv')
DPI_prices = DPI_prices.rename(columns={'day': 'Date'})
DPI_prices = DPI_prices.loc[DPI_prices['symbol'] == 'DPI'] 
DPI_prices['Date'] = DPI_prices['Date'].astype('datetime64[ns]') 
DPI_prices = DPI_prices.drop(columns='symbol')
DPI_prices.set_index('Date',inplace=True)
```

```python
underlying_assets = pd.read_csv(path_to_module + 'Underlying Assets Data.csv')
underlying_assets = underlying_assets.rename(columns={'day': 'Date'})
underlying_assets['Date'] = underlying_assets['Date'].astype('datetime64[ns]') 
underlying_assets.set_index('Date',inplace=True)
underlying_assets = underlying_assets.pivot(columns='symbol',values='price')
underlying_assets = underlying_assets.rename_axis(None, axis=1).reset_index()
underlying_assets = underlying_assets[['Date','UNI','AAVE','MKR','SNX','COMP']]
```

```python
benchmarks = pd.read_csv(path_to_module + 'Benchmark_Prices.csv')
benchmarks = benchmarks.rename(columns={'day': 'Date'})
benchmarks['Date'] = benchmarks['Date'].astype('datetime64[ns]') 
benchmarks.set_index('Date',inplace=True)
benchmarks = benchmarks.pivot(columns='symbol',values='price')
benchmarks = benchmarks.rename_axis(None, axis=1).reset_index()
```

```python
market_data = pd.read_csv(path_to_module + '/Market_Data_cleaned_no_commas.csv')
market_data['Date'] = market_data['Date'].astype('datetime64[ns]') 
market_data.set_index('Date',inplace=True)
market_data = market_data.reindex(pd.date_range(market_data.index.min(), market_data.index.max())).sort_index(ascending=True).reset_index().rename(columns={'index': 'Date'})
market_data = market_data.fillna(method='ffill')
```

```python
sofr = pd.read_csv(path_to_module + '/SOFR_historical.csv')
sofr.rename(columns={'DATE': 'Date'}, inplace=True)
sofr['Date'] = sofr['Date'].astype('datetime64[ns]') 
sofr.set_index('Date',inplace=True)
sofr = sofr.reindex(pd.date_range(sofr.index.min(), sofr.index.max())).sort_index(ascending=True).reset_index().rename(columns={'index': 'Date'})
sofr = sofr.fillna(method='ffill')
# sofr['SOFR'] = sofr['SOFR'].astype('float64') 
sofr['SOFR'] = pd.to_numeric(sofr['SOFR'], errors ='coerce')
```

```python
liquidity = pd.read_csv(path + '/UniV3 TVL.csv')
liquidity.rename(columns={'day': 'Date'}, inplace=True)
liquidity['Date'] = liquidity['Date'].astype('datetime64[ns]') 
liquidity.set_index('Date',inplace=True)
liquidity = liquidity.reindex(pd.date_range(liquidity.index.min(), liquidity.index.max())).sort_index(ascending=True).reset_index().rename(columns={'index': 'Date'})
liquidity = liquidity.fillna(method='ffill')
liquidity.drop(columns='ma_tvl',inplace=True)
liquidity
# sofr['SOFR'] = sofr['SOFR'].astype('float64') 
# liquidity['SOFR'] = pd.to_numeric(sofr['SOFR'], errors ='coerce')
```

```python
trading_vol = pd.read_csv(path + '/DPI Trading Vol.csv')
trading_vol.rename(columns={'day': 'Date'}, inplace=True)
trading_vol['Date'] = trading_vol['Date'].astype('datetime64[ns]') 
trading_vol.set_index('Date',inplace=True)
trading_vol = trading_vol.reindex(pd.date_range(trading_vol.index.min(), trading_vol.index.max())).sort_index(ascending=True).reset_index().rename(columns={'index': 'Date'})
trading_vol = trading_vol.fillna(method='ffill')
trading_vol.drop(columns='product',inplace=True)
trading_vol
```

```python
inflows = pd.read_csv(path_to_module + '/Net_inflow_and_aum.csv')
inflows = inflows.loc[inflows['symbol'] == 'DPI'] 
inflows = inflows.rename(columns={'day': 'Date'})
inflows['Date'] = inflows['Date'].astype('datetime64[ns]') 
inflows = inflows.drop(columns=['symbol', 'gross_flow'])
inflows.set_index('Date',inplace=True)
```

```python
# importing fear_index
fear_index = pd.read_csv( '/content/gdrive/My Drive/IndexCoop Capstone/DPI analysis/datasets/crypto_fear_index_historical.csv')
fear_index['Date'] = fear_index['Date'].astype('datetime64[ns]') 
fear_index.set_index('Date', inplace=True)
fear_index
```

```python
final_df= DPI_prices.merge(underlying_assets, on='Date').merge(benchmarks,on="Date").merge(inflows,on="Date").merge(market_data,on="Date").merge(sofr,on="Date").merge(fear_index, on="Date").merge(trading_vol,on="Date")
final_df.set_index('Date',inplace=True)
final_df
```
