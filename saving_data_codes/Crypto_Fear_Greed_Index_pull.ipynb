```python
#  def function for kinda "standardizing" fear index value classification 

def fear_index_classification(data):
  data.loc[data['value_classification'] == 'Extreme Fear', 'fear_index'] = '-2'
  data.loc[data['value_classification'] == 'Fear', 'fear_index'] = '-1'
  data.loc[data['value_classification'] == 'Neutral', 'fear_index'] = '0'
  data.loc[data['value_classification'] == 'Greed', 'fear_index'] = '1'
  data.loc[data['value_classification'] == 'Extreme Greed', 'fear_index'] = '2'
```

```python
# requesting data from alternative.me API 
r = requests.get('https://api.alternative.me/fng/?limit=0') 
```

```python
df = pd.DataFrame(r.json()['data'])
df.value = df.value.astype(int) 

# this timestamp unit may not be the best one, I think can be changed as needed
df.timestamp = pd.to_datetime(df.timestamp, unit='s')
df = df.rename(columns = {'timestamp' : 'Date'})
df.set_index('Date', inplace=True) 

# sorting date from min to max 
df = df[::-1] 

# re-classifying fear index 
fear_index_classification(df)


# # removing columns that are not that necessary anymore
df = df.drop(['time_until_update', 'value', 'value_classification'], axis=1) 
df.to_csv('/content/gdrive/My Drive/IndexCoop Capstone/DPI analysis/datasets/crypto_fear_index_historical.csv') 
```
