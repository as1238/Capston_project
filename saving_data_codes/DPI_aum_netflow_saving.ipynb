```python
DPI_df = pd.read_csv('/content/gdrive/My Drive/IndexCoop Capstone/CSV data files/Historical_Token_Prices.csv') 
DPI_df 
```

```python
DPI_df = DPI_df.rename(columns={'day': 'Date'})
DPI_df = DPI_df.rename(columns={'price': 'DPI_price'})
```

```python
DPI_df
```

```python
DPI = DPI_df.loc[DPI_df['symbol'] == 'DPI'] 
DPI
```

```python
DPI['Date'] = DPI['Date'].astype('datetime64[ns]') 
DPI 
```

```python
DPI = DPI.set_index(['Date']) 
DPI
```

```python
# DPI['DPI_returns'] = DPI['price'].pct_change() 
DPI = DPI.drop(columns=['symbol']) # , 'price'
DPI 
```

```python
# DPI.to_csv('/content/gdrive/My Drive/IndexCoop Capstone/DPI analysis/datasets/DPI_returns.csv')   ## no need for this, as we'll combine them two 
```

```python
main_AUM = pd.read_csv('/content/gdrive/My Drive/IndexCoop Capstone/CSV data files/Net_inflow_and_aum.csv') 

Net_AUM = main_AUM.loc[main_AUM['symbol'] == 'DPI'] 

# Net_AUM['aum_change'] = Net_AUM['aum'].pct_change()  ## trying without change 
Net_AUM 

```

```python
# Net_AUM['net_flow_change'] = Net_AUM['net_flow'].pct_change() ## trying without this 
Net_AUM = Net_AUM.drop(columns=['symbol', 'gross_flow'])  ## removing this 'aum', 
Net_AUM = Net_AUM.rename(columns={'day': 'Date'})
Net_AUM['Date'] = Net_AUM['Date'].astype('datetime64[ns]') 
Net_AUM = Net_AUM.set_index(['Date']) 
Net_AUM 
```

```python
DPI_aum_netflow = pd.DataFrame()

DPI_aum_netflow = pd.merge(Net_AUM, DPI, left_index=True, right_index=True) 
DPI_aum_netflow 

# pd.concat([df, df1], axis=0, ignore_index=True)

# pd.merge(df1, df2, left_index=True, right_index=True)

```
